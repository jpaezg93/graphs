```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import numpy as np
```

```python
data = '''id	materia	nota	aprobado
1	Matemáticas	80	Sí
2	Historia	65	No
3	Ciencias	90	Sí
4	Lenguaje	75	Sí
5	Matemáticas	95	Sí
6	Historia	70	Sí
7	Ciencias	85	Sí
8	Lenguaje	60	No
9	Matemáticas	78	Sí
10	Historia	82	Sí
11	Ciencias	93	Sí
12	Lenguaje	68	Sí
13	Matemáticas	73	Sí
14	Historia	88	Sí
15	Ciencias	77	Sí
16	Lenguaje	50	No
17	Matemáticas	92	Sí
18	Historia	63	No
19	Ciencias	85	Sí
20	Lenguaje	79	Sí'''
col1 = []
col2 = []
col3 = []
col4 = []
data = data.split('\n')
for row in data:
    row = row.split()
    if row[0]=='id':
        keys = row
    else:
        col1.append(row[0])
        col2.append(row[1])
        col3.append(row[2])
        col4.append(row[3])

data = {keys[0]:col1, keys[1]:col2, keys[2]:col3, keys[3]:col4,}
df = pd.DataFrame(data)
df['nota'] = df['nota'].astype('int64')
df['materia'] = df['materia'].astype('str')
df['aprobado'] = df['aprobado'].astype('str')
df
```

```python
plt.style.use('fivethirtyeight')
box_trace = go.Box(
    x=df.materia,
    y=df.nota,
    name='notas'
)

# Create figure and add trace
fig = go.Figure(box_trace)

# Update layout
fig.update_layout(title='grafico de cajas', boxmode='group')

# Display the figure
fig.show()
```

```python
group = df.groupby('aprobado').size()

fig = go.Figure()
# Añadir el trazado de torta a la figura
fig.add_trace(go.Pie(
    labels=list(group.index), values=list(group.values),  # Etiquetas y valores
    hole=0  # Tamaño del hueco en el centro del gráfico de torta
))

# Personalizar el diseño del gráfico
fig.update_layout(
    title='Gráfico de Torta'  # Título del gráfico
)

# Mostrar el gráfico
fig.show()
```
