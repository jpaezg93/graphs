```python
import numpy as np
import matplotlib.pyplot as plt

rng = np.random.default_rng(42)

# Datos de calificaciones de los estudiantes
matematicas = rng.integers(50, 100, 20)
ciencias = rng.integers(40, 95, 20)
literatura = rng.integers(60, 100, 20)

# Datos de errores para el gr치fico de barras de error
errores_matematicas = rng.uniform(2, 5, 2)
errores_matematicas = [min(errores_matematicas), max(errores_matematicas)]

errores_ciencias = rng.uniform(1, 4, 2)
errores_ciencias = [min(errores_ciencias), max(errores_ciencias)]

errores_literatura = rng.uniform(3, 6, 2)
errores_literatura = [min(errores_literatura), max(errores_literatura)]

```

```python
plt.scatter(matematicas, ciencias, marker='o', linestyle='-', color='blue')
plt.title('Relaci칩n entre las calificaciones de Matem치ticas y Ciencias')
plt.xlabel('Calificaciones de Matem치ticas')
plt.ylabel('Calificaciones de Ciencias')

plt.show()
```

```python
materias = ['Matem치ticas', 'Ciencias', 'Literatura']
errores = np.array([
    [min(errores_matematicas), min(errores_ciencias), min(errores_literatura)],
    [max(errores_matematicas), max(errores_ciencias), max(errores_literatura)]
])
plt.errorbar(x = materias, y = [np.mean(matematicas), np.mean(ciencias), np.mean(literatura)],yerr=errores , fmt='o', capsize=5)

plt.title('Calificaciones Promedio por Materia con Barras de Error')
plt.ylabel('calificaciones promedio')

plt.show()
```

```python
plt.hist(matematicas, bins=10, edgecolor='black')
plt.title('Distribuci칩n de las calificaciones de Matem치ticas')
plt.xlabel('Calificaciones de Matem치ticas')
plt.ylabel('Frecuencia')

plt.show()
```
